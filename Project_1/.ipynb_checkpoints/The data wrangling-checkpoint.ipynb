```python
#import relevant libraries for the data analysis 
import pandas as pd 
import numpy as np
print('libraries imported')
pd.set_option('display.max_columns', None)
```

```python
#import the dataset
website_url='https://s3.us.cloud-object-storage.appdomain.cloud/cf-courses-data/CognitiveClass/DP0701EN/version-2/Data-Collisions.csv'
Df= pd.read_csv(website_url, sep=',', error_bad_lines=False, index_col=False,dtype='unicode')
Df.head()
```

```python
Df.tail()
```

```python
Df['X']=Df['X'].astype('float')
Df['Y']=Df['Y'].astype('float')
Df['SEVERITYCODE']=Df['SEVERITYCODE'].astype('int')
Df['OBJECTID']=Df['OBJECTID'].astype('int')
Df['INCKEY']=Df['INCKEY'].astype('int')
Df['COLDETKEY']=Df['COLDETKEY'].astype('int')
Df['REPORTNO']=Df['REPORTNO'].astype('str')
Df['STATUS']=Df['STATUS'].astype('str')
Df['ADDRTYPE']=Df['ADDRTYPE'].astype('str')
Df['INTKEY']=Df['INTKEY'].astype('float')
Df['LOCATION']=Df['LOCATION'].astype('str')
Df['SEVERITYDESC']=Df['SEVERITYDESC'].astype('str')
Df['COLLISIONTYPE']=Df['COLLISIONTYPE'].astype('str')
Df['PERSONCOUNT']=Df['PERSONCOUNT'].astype('int')
Df['PEDCOUNT']=Df['PEDCOUNT'].astype('int')
Df['PEDCYLCOUNT']=Df['PEDCYLCOUNT'].astype('int')
Df['VEHCOUNT']=Df['VEHCOUNT'].astype('int')
Df['JUNCTIONTYPE']=Df['JUNCTIONTYPE'].astype('str')
Df['INATTENTIONIND']=Df['INATTENTIONIND'].astype('str')
Df['UNDERINFL']=Df['UNDERINFL'].astype('str')
Df['WEATHER']=Df['WEATHER'].astype('str')
Df['ROADCOND']=Df['ROADCOND'].astype('str')
Df['LIGHTCOND']=Df['LIGHTCOND'].astype('str')
Df['PEDROWNOTGRNT']=Df['PEDROWNOTGRNT'].astype('str')
Df['SPEEDING']=Df['SPEEDING'].astype('str')
```
